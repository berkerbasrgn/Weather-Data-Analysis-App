```python
import pandas as pd

weather_df = pd.read_csv("/kaggle/input/weather/weather_data.csv")

weather_df.head()
```

```python
import numpy as np

sample_size = weather_df["Temperature (°C)"].shape[0]

random_variables = np.random.uniform(low=0, high=3, size=sample_size)

random_variables
```

```python
weather_df["temp_add"] = random_variables

weather_df.head(20)
```

```python
weather_df["temp_increased"] = weather_df["Temperature (°C)"] + weather_df["temp_add"]

weather_df.head(20)
```

```python
a = weather_df["Temperature (°C)"].describe()
b = weather_df["temp_increased"].describe()

print("temperature:", a)
print("tempincreasewed", b)

```

```python
weather_df.set_index("Date", inplace=True)

weather_df.head()
```

```python
weather_df.columns
```

```python
weather_df.index = pd.to_datetime(weather_df.index)

months = weather_df.index.month

avgmonth = weather_df.groupby(months)["Temperature (°C)"].mean()

result = avgmonth[avgmonth > 20]

result
```

```python
import sqlite3
import pandas as pd

database_file = "weather_data.db"
table_name = "weather"

weather_df.to_sql(table_name, con=sqlite3.connect(database_file), if_exists="replace", index=False)

print(f"Data exported to SQLite database: {database_file}")

conn = sqlite3.connect(database_file)
cursor = conn.cursor()

cursor.execute(f"SELECT * FROM {table_name}")
rows = cursor.fetchall()

column_names = [description[0] for description in cursor.description]
print("\nColumn Names:", column_names)
print("\nData in the Database:")
for row in rows:
    print(row)

conn.close()
```

```python
import pandas as pd
import plotly.express as px

weather_df.index = pd.to_datetime(weather_df.index)

weather_df["Month"] = weather_df.index.month

monthly_means = weather_df.groupby('Month')[['Temperature (°C)', 'temp_increased']].mean().reset_index()

print("Monthly Mean Temperatures:")
print(monthly_means)

fig = px.line(monthly_means, x='Month', y=['Temperature (°C)', 'temp_increased'],
              title='Monthly Mean Temperature vs. Temperature Increased',
              labels={'value': 'Temperature (°C)', 'variable': 'Column'},
              markers=True)

fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Temperature (°C)',
    legend_title='Column',
    xaxis=dict(tickmode='linear', tick0=1, dtick=1),  
    template='plotly_white'
)

fig.show()
```
