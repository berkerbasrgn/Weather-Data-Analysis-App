```python
import sqlite3
import pandas as pd

database_file = "weather_data.db"
table_name = "weather"

weather_df.to_sql(table_name, con=sqlite3.connect(database_file), if_exists="replace", index=False)

print(f"Data exported to SQLite database: {database_file}")

conn = sqlite3.connect(database_file)
cursor = conn.cursor()

cursor.execute(f"SELECT * FROM {table_name}")
rows = cursor.fetchall()

column_names = [description[0] for description in cursor.description]
print("\nColumn Names:", column_names)
print("\nData in the Database:")
for row in rows:
    print(row)

conn.close()
```

```python
import pandas as pd
import plotly.express as px

weather_df.index = pd.to_datetime(weather_df.index)

weather_df["Month"] = weather_df.index.month

monthly_means = weather_df.groupby('Month')[['Temperature (째C)', 'temp_increased']].mean().reset_index()

print("Monthly Mean Temperatures:")
print(monthly_means)

fig = px.line(monthly_means, x='Month', y=['Temperature (째C)', 'temp_increased'],
              title='Monthly Mean Temperature vs. Temperature Increased',
              labels={'value': 'Temperature (째C)', 'variable': 'Column'},
              markers=True)

fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Temperature (째C)',
    legend_title='Column',
    xaxis=dict(tickmode='linear', tick0=1, dtick=1),  
    template='plotly_white'
)

fig.show()
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px

# I USED GEMINI FOR CREATING THESE VARIABLES:
data = {
    'Date': pd.date_range(start='2023-01-01', periods=365, freq='D'),
    'Temperature (째C)': [15 + 10 * (i % 30) / 30 for i in range(365)],
}


df = pd.DataFrame(data)
df.set_index('Date', inplace=True)
df['temp_increased'] = df['Temperature (째C)'] + 2  # Simple increase of 2째C
df['Month'] = df.index.month



monthly_means = df.groupby('Month')[['Temperature (째C)', 'temp_increased']].mean().reset_index()


# CPT IS USED FOR PLOTLY LIBRARY

fig = px.line(monthly_means, x='Month', y=['Temperature (째C)', 'temp_increased'],
              title='Monthly Mean Temperature vs. Temperature Increased',
              labels={'value': 'Temperature (째C)', 'variable': 'Column'},
              markers=True)
fig.update_layout(xaxis_title='Month', yaxis_title='Temperature (째C)', template='plotly_white')


# I USED MY PYTHON PROJECT FOR THESE CODES:

st.title("weather data analysis")
st.header("orignal data")
st.write(df)
st.header("monthly mean temmp")
st.write(monthly_means)
st.header("monthly mean temp vstemp increased")
st.plotly_chart(fig)
```
